# Matplotlib
Matplotlib is the most popular plotting library for Python.
It give you control over every aspect of a figure.
It was Designed to have a similar feel to MatLanb's Graphical Plotting.
If you wwant to work on Matplotlib we have to install it by giving command pip install matplotlib or conda install matplotlib.
THere are 2 ways to create a plots or charts by using Matplotlib. Functional and Object Oriented. In our this notebook both way is going to be discussed.

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt #importing the library
%matplotlib inline
```

```python
x = np.linspace(0,5,11)
y = x**2
x,y
```

```python
#functional way
plt.plot(x,y)
plt.show()
```

```python
plt.plot(x,y)
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Aman')
plt.show()
```

```python
#chnage the colour
plt.plot(x,y,'r')
plt.xlabel('X-axis')
plt.ylabel('y-axis')
plt.title('Aman')
plt.show()
```

```python
#MultiPlot in the same canvas(1,2,1- one row 2 column and plot number)
plt.subplot(1 ,2,1)
plt.plot(x,y,'r')

plt.subplot(1 ,2,2)
plt.plot(y,x,'b')
```

```python
#object orineted way of ploting 
fig = plt.figure() #create a object of Figure class that you can consider an imaginary blank canvas 
axes = fig.add_axes([0.1,0.1,0.8,0.8])#list consist of length , bottom, widht and height that range from 0 to 1
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**4)
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2)
axes.plot(x,x**4)
axes.legend()
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = "x square")
axes.plot(x,x**3 , label = "X cube")
axes.legend()
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = "x square",color='purple',linewidth=3,alpha=2)
#alpha is for how much line transparent is 
axes.plot(x,x**3 , label = "X cube")
axes.legend()
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = "x square",color='purple',lw=3,linestyle='--')
#alpha is for how much line transparent is 
axes.plot(x,x**3 , label = "X cube")
axes.legend()
```

```python
fig = plt.figure(figsize=(5,4))
axes = fig.add_axes([0,0,1,1])
axes.plot(x,x**2,label = "x square",color='purple',linewidth=3,alpha=2,linestyle='steps')
#alpha is for how much line transparent is 
axes.plot(x,x**3 , label = "X cube")
axes.legend()
```
