# Bar Chart(count)

```python
def rank_performer(value):
    if value<=200:
        return "Less Than and equal to 200"
    elif value >200 and value<=500:
        return "Greater than 200 and less than or equal to 500 "
    else:
        return "More than 500"
    
a = data['Close'].apply(rank_performer).value_counts()
```

```python
a.plot(kind='bar')
```

```python
a.plot(kind='barh')
```

```python
data['Close'].mean()
```

```python
def rank_checker(value):
    if value<327.7248241041954:
        return "Below Average"
    else:
        return "Above Average"
data['Close'].apply(rank_checker).value_counts().plot(kind='pie')
```

```python
def rank_checker(value):
    if value<327.7248241041954:
        return "Below Average"
    else:
        return "Above Average"
data['Close'].apply(rank_checker).value_counts().plot(kind='pie',legend = True)
```

```python
def buckets(value):
    return int(value/100.0)*100

data['High'].apply(buckets)
```
