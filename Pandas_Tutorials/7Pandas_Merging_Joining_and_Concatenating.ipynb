```python
"""
append mathod , which basically some how like the concat function but the 
diffrence between these two is contact is function which is directly call by 
pandas ans append method is directly work on dataframe

"""
sales = week1.append(week2, ignore_index=True)
sales
```

```python
#inner join : it merge 2 dataset based on commanly shared values 
week1.merge(week2)
```

```python
#lets expand more about this merge method
#week1 as left data week2 as right , how = innner mean inner join and on to which column
s = week1.merge(week2 , how="inner",on ="Customer ID")#foodx will be food id of left table 
```

```python
#here you can see the some customer id is reapeating
s.head() #155 and 503 is reapeting
week1[week1['Customer ID']==155] 
#here you cam see this order id comes 2 times with diffrent food order
```

```python
s
#instead of x and y we can configure the suffix as well.
s = week1.merge(week2 , how="inner",on ="Customer ID", suffixes=["-Week1","-Week2"])
s
```

```python
"""the last we had perform inner join on one column that is customer id
that who are comming on the both week. Now i want that customer who come 
on week1 and week2 and order the same food. That means we are implementing 
inner join on multiple columns
"""
week1.merge(week2,on=['Customer ID','Food ID'])#21 and 578 are repeating
```

```python
#outer Join: which combine value that are found in either dataframe or both
```

```python
week1.merge(week2,how="outer",on='Customer ID',suffixes=[' -Week1',' -Week2'])
```

```python
merged = week1.merge(week2,how="outer",on='Customer ID',suffixes=[' -Week1',' -Week2'],
           indicator=True)
```

```python
merged['_merge'].value_counts()
# we got an proper idea which how much customer is in df1 , df2 and both
```

```python
a = merged["_merge"].isin(["left_only",'right_only'])
# to remove the intersection portion
merged[a]
```

```python
#Left Join: all the worws which rows are avaible in left table and intersect
week1.merge(food,how = "left", on ='Food ID',sort=True)
```

```python
"""
Till now we had work on inner , outer and left join. It was working fine
because we have the same column name in the both datasets.
But if we dont have any same name then.
In this situation pandas merge method provide us 2 more parameters.
left on and right on. Lets have a look on  example
"""
week1.head()#it consist of customer id , food id.
customer.head() #it consist of Id , first name and so on 
```

```python
week1.merge(customer,how="left",left_on='Customer ID',right_on='ID')
```

```python
#droping the repating column
week1.merge(customer,how="left",left_on='Customer ID',right_on='ID').drop('ID',axis='columns')
```

```python
#merginmg on the basis of index(take the left and right index and merge it)
```

```python
customer1 = pd.read_csv("datasets/Restaurant - Customers.csv",index_col='ID')
food1 = pd.read_csv("datasets/Restaurant - Foods.csv",index_col='Food ID')
```

```python
#merging wee1 on column name as Customer ID and on customer by index
```

```python
sales1 = week1.merge(customer1,how="left",left_on='Customer ID' , right_index=True)
```

```python
sales1.merge(food1,how='left',left_on='Food ID',right_index= True)
```

```python
#merging on the basis of index by which i get to know which customer is one on week1 and week2
week1.merge(week2,how='left',left_index=True,right_index=True)
```

```python
#join method : whih concat the dataframe vertically when two dataframe share 
#the exact same index. it is some how similar to append and concat but on the diffrent axsis
```

```python
satisfaction = pd.read_csv('datasets/Restaurant - Week 1 Satisfaction.csv')
satisfaction
```

```python
#get the satisfaction of the customer by week1 sales data
#one way
week1.merge(satisfaction,how="left",left_index=True , right_index=True)
```

```python
#another wat
week1.join(satisfaction)#join base on index by default
```
