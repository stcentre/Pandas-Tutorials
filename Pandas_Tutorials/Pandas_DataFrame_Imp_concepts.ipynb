```python
import pandas as pd
```

```python
fortune = pd.read_csv('Pandas_Tutorials/datasets/fortune1000.csv',index_col='Rank')
fortune
```

```python
"""So ideal thing you want to a groupby is always going to be a
column that has duplicate values.This basically means you want to create
segments of you dataset.SO in our this dataset there is not any
point to grouping on company because every single valeu of the company is 
unique but on grouping on the column name as sector make some sense
Similar we can perfrom this concept to industry column as well.
The froup by objcet is created directly from dataframe and the syntax is"""
#return a me datafragroup by object
sector = fortune.groupby('Sector')
type(fortune)
type(sector)
#a group by object is pointless untill we call methods upon it
```

```python
#return the len of the dataframe groupby object
len(sector)
#return the unique number of values in object
fortune['Sector'].nunique()
#there are 21 diffrenct sector in 1000 rows of dataset
```

```python
"""
It will return a series which consist of name of the sector as 
the index and total occurence of that sector in the dataset as 
value of the series
"""
sector.size()
#same 
fortune['Sector'].value_counts()
#return the first occurence of the value of the sector where grouping will be index label
sector.first()
#return the first occurence of the value of yhe sector where grouping will be index label
sector.last()
#the convenience of the first and last method is to able to
#extract just a sample rows from very last and very first
```
