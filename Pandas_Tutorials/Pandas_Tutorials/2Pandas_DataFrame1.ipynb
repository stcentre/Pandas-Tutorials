```python
nba = pd.read_csv("Pandas_Tutorials/datasets/nba.csv")
```

```python
#filling the missing values
nba.fillna(0)
nba['College'].fillna("No COllage")
```

```python
#astype which allow to convert the datatype of series to one another
nba['Salary'].fillna(0,inplace=True)
nba['College'].fillna('None',inplace=True)
nba.head()
nba.info()
nba['Salary']=nba['Salary'].astype("int")
nba.info()
```

```python
#retrun that how much unique values are there in dataframe
nba['Position'].nunique()
#converting from string to catpgory
nba['Position']=nba['Position'].astype("category")
nba.dtypes
```

```python
#sorting the datafame by values
#remember one thing this method require one argument that is column name 
#that make sense on which column you want to sort it 
nba.sort_values('Salary')
#another paramerter we have name as ascending default argument is True
#return the df on desceding value of salary
nba.sort_values('Salary',ascending=False)
#return datafrane nan value will be on the top
nba.sort_values('Salary',na_position='first')
```

```python
#sort by multiple columns
#it will return first it sorts by team and after that with in each team
#it moves to the player and then its sort the name 
nba.sort_values(['Team','Name'])
#now lets make som of the changes in ascending or descding
#team in asceding and name in desceding
nba.sort_values(['Team','Name'],ascending=[True,False])
```
