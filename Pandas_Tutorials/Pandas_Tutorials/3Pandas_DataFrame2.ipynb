```python
#Retrive rows by index label with .loc
#if you are using the string  the index thats our convention says
bond.sort_index(inplace=True)
```

```python
"""Now we are going to use loc method which directly work with dataframe
and this is a method which is not using paranthesis but using square 
braces.Basically its a method which is used for extraction using the 
index"""
#and this will return the series of the values if you are passing single index
bond.loc['Goldfinger']
#if the index or film in our case is not availble it will throw an error
# bond.loc['abc']
#multiple indedx left one is starting last one is ending
bond.loc['Diamonds Are Forever':'Moonraker']
"""Here you can see out an diffrence 
whenever we are we are extracting in the list our last item is exclusive 
that means its not returnig that item but when ever you are using an string
index the last item is included"""
#all the movie but the index start from GoldeEye
bond.loc['GoldenEye':]
#multiple index
bond.loc[['GoldenEye','Moonraker']]
```

```python
#Retrive rows by index position with iloc
#index position rather than the index label
#this make sense either we have index label we can extract by giving default index
bond.head(5)
#return first five rows
bond.iloc[:5]
#same you can use the indexing concept
#this will return the value which is on 5th index and 4th column
bond.iloc[5,4]
#returning multiple rows of one one given column that is 4th
bond.iloc[:10,4]
#for rows we have to write on the left hand side of ',' and for columns 
#we have to mention the column index at the right
```

```python
#the ix() method
#this a optimal method when you want to mix and match index label and index Positions
#return the row of given index
bond.ix['GoldenEye']
bond.ix[['Moonraker','GoldenEye']]
#here you can see out index position is also be executed in ix
bond.ix[10]
```

```python
#The second arguments in loc , iloc and ix method
#second argument in loc is the name of columns you want
bond.loc["GoldenEye",['Actor','Box Office']]
#in same way second argument in iloc will be the column position
bond.iloc[5,[1,5]]
#but in ix we can pass either the name of the column either the position of th column
bond.ix[5,'Actor']
bond.ix[5:10,['Actor','Box Office']]
```

```python
#set a new value for a an existing value in specific column
#return the actor of the movie name as GoldenEye
bond.ix['GoldenEye','Actor']
#this is i am changing the name of the actor
bond.ix['GoldenEye','Actor']="ABCD"
bond.ix['GoldenEye','Actor']
#chnaging the multiple columns
bond.ix['GoldenEye',['Box Office','Budget']]
bond.ix['GoldenEye',['Box Office','Budget']]=[10,20]
bond.ix['GoldenEye',['Box Office','Budget']]
```
