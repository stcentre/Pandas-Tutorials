```python
"""
The Ideal thing that you want to groupby is always going to be a column that has duplicate value.
This basically mean you want to create a segments of your dataset.For Example there is not any 
point that we are grouping on company because every single vale in company column is unique.
But if we are grouping on sector its make some how sense because companies belong to a sector and
we can see how much unique value is availble in Sector. In sector only 
21 unqiue values are there in 1000 nrecord so we can make group on secotor.
So Sector is exactly perfect for us to Group by and the reason is that because there are multiple 
companies that fall into the exact same sector.Similary Industry is another good column because
multiple companies will fall to same industry as well.

lets have a look on syntax how we can use group by in on sector.

"""
data['Sector'].nunique()
```

```python
"""
Sytax is we have to call a method name as groupby which take argument as a column name
on which you want to make a groups.So how is this going to work behind the sence.
Pandas is going to take a look at all of the values in the sector column for example its going 
to start at retail. Its then going to look through the data set and collect all the rows that fall
onto that sector.So after executing the follwing command this is going to be printed.
It will create a pandas dataframe grouo by object.
"""

data.groupby('Sector')
```

```python
#now data is dataframe object and sector is groupby object. These 2 are seprtely object of pandas
#you can see
sector = data.groupby('Sector')
print(type(data)) #'pandas.core.frame.DataFrame'
print(type(sector)) #'pandas.core.groupby.generic.DataFrameGroupBy'
```
